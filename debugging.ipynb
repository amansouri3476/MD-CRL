```python
import torch
import numpy
import os
```

```python
# epochs_list = [34, 35, 36, 37]
# epochs_list = list(range(40,60))
epochs_list = list(range(10))

# no BN: /home/mila/s/sayed.mansouri-tehrani/scratch/logs/training/multiruns/autoencoder_balls_encoded_128/2023-10-05_13-50-15/model.hinge_loss_weight=0.0,model.mmd_loss.fix_sigma=null,model.mmd_loss.kernel_multiplier=1.0,model.mmd_loss.kernel_number=3,model.penalty_weight=10.0,model.save_encoded_data=True,model.z_dim_invariant_fraction=0.8,model/autoencoder=mlp_ae_balls,model=balls_md_encoded_autoencoder/
path_header = "/home/mila/s/sayed.mansouri-tehrani/scratch/logs/training/multiruns/autoencoder_balls_encoded_128/2023-10-05_16-09-57/model.hinge_loss_weight=0.0,model.penalty_weight=1.0,model.save_encoded_data=True,model.z_dim_invariant_fraction=0.8,model/autoencoder=mlp_ae_balls,model=balls_md_encoded_autoencoder/"

train_dataset_list = []
valid_dataset_list = []

for epoch in epochs_list:
    train_dataset_list.append(torch.load(os.path.join(path_header, f"double_encoded_img_balls_encoded_train_epoch_{epoch}.pt")))
    valid_dataset_list.append(torch.load(os.path.join(path_header, f"double_encoded_img_balls_encoded_valid_epoch_{epoch}.pt")))
```

```python
from sklearn.linear_model import LinearRegression


z_dim_invariant = 102

for i, train_dataset in enumerate(train_dataset_list):
    print(f"================================================================================================")
    z_hat = train_dataset["z_hat"]
    z_hat_inv = train_dataset["z_hat"][:, :z_dim_invariant]
    z_inv = train_dataset["z_invariant"]
    z_spu = train_dataset["z_spurious"]
    z = train_dataset["z"]

    print(f"z_hat.shape: {z_hat.shape}, z_hat_inv.shape: {z_hat_inv.shape}, z_inv.shape: {z_inv.shape}, z_spu.shape: {z_spu.shape}, z.shape: {z.shape}")

    reg = LinearRegression().fit(z_hat, z)
    r2 = reg.score(z_hat, z)
    print(f"R2 score for z_hat predicting z: {r2}")
    reg = LinearRegression().fit(z, z_hat)
    r2 = reg.score(z, z_hat)
    print(f"R2 score for z predicting z_hat: {r2}")

    reg = LinearRegression().fit(z_hat_inv, z_inv)
    r2 = reg.score(z_hat_inv, z_inv)
    print(f"R2 score for z_hat_inv predicting z_inv: {r2}")

    reg = LinearRegression().fit(z_hat_inv, z_spu)
    r2 = reg.score(z_hat_inv, z_spu)
    print(f"R2 score for z_hat_inv predicting z_spu: {r2}")

    reg = LinearRegression().fit(z_inv, z_hat_inv)
    r2 = reg.score(z_inv, z_hat_inv)
    print(f"R2 score for z_inv predicting z_hat_inv: {r2}")

    reg = LinearRegression().fit(z_spu, z_hat_inv)
    r2 = reg.score(z_spu, z_hat_inv)
    print(f"R2 score for z_spu predicting z_hat_inv: {r2}")
```
